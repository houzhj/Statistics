## (1) Beta Distribution

**In Bayesian inference, the beta distribution is the conjugate prior probability distribution for the Bernoulli, binomial, negative binomial and geometric distributions.**
- Beta distribution [[Reference]](https://en.wikipedia.org/wiki/Beta_distribution)
- Conjugate prior probability distribution: The posterior distribution is in the same probability distribution family as the prior probability distribution.[[Reference]](https://en.wikipedia.org/wiki/Conjugate_prior)

**The following plots show the pdf for Beta distribution with different combinations of parameters.**

```python
def create_lable(alpha,beta):
    return 'Beta('+str(alpha)+','+str(beta)+')'
```

```python
x = np.linspace(0, 1, 1000)
plt.plot(x, beta.pdf(x, 2, 5), label=create_lable(2,5))
plt.plot(x, beta.pdf(x, 5, 2), label=create_lable(5,2))
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 1, 1000)
for i in [i*2 for i in range(1,6)]:
    plt.plot(x,beta.pdf(x,1,i),label=create_lable(1,i))
plt.legend(loc='upper right', bbox_to_anchor=(1.3,1))
plt.title('PDF of Beta(1,x)')
plt.show()
```

```python
x = np.linspace(0, 1, 1000)
for i in [i*2 for i in range(1,6)]:
    plt.plot(x,beta.pdf(x,2,i),label=create_lable(2,i))
plt.legend(loc='upper right', bbox_to_anchor=(1.3,1))
plt.show()
```

```python
x = np.linspace(0, 1, 1000)
for i in [0.5]+[i*2+1 for i in range(5)]:
    plt.plot(x,beta.pdf(x,i,i),label=create_lable(i,i))
plt.legend(loc='upper right', bbox_to_anchor=(1.3,1))
plt.title('PDF of Beta(x,x)')
plt.show()
```

## (2) Bayes estimate of Bernoulli(p) with Beta prior distribution

### [See derivations here](https://github.com/houzhj/Statistics/blob/main/math/01_bayes_estimate_derivation_01.pdf)

```python
def Bayesian_ber_prior_beta(true_p,sample_size,alpha_prior,beta_prior,plot):
    sample = np.random.binomial(1,true_p,sample_size)
    sample.mean()
    
    alpha_posterior = alpha_prior + sample.sum()
    beta_posterior  = beta_prior + len(sample) - sample.sum()
    
    mean_prior     = round(alpha_prior/(alpha_prior+beta_prior),3)
    mean_posterior = round(alpha_posterior/(alpha_posterior+beta_posterior),3)
    
    x = np.linspace(0,1,1000)
    
    if plot == True:
        plt.plot(x,beta.pdf(x,alpha_prior,beta_prior),color='blue',
                 label='prior distribution:'+create_lable(alpha_prior,beta_prior))

        plt.plot(x,beta.pdf(x,alpha_posterior,beta_posterior),color='red',
                 label='posterior distribution:'+create_lable(alpha_posterior,beta_posterior))
        plt.axvline(x=mean_posterior, color='red', linestyle='--', label='Bayes estimate')
        plt.axvline(x=sample.mean(), color='green', linestyle='--', label='MLE')

        plt.axvline(x=true_p, color='black', linestyle='--', label='true value')
        plt.legend()
        plt.show()

    return(mean_posterior,round(sample.mean(),3))
```

### Try a few examples of priors, consider the two aspectsï¼š
- **Correct**
- **Informative** (The concept of "informative" here is relative, rather than strictly defined. There are formal definitions for [informative prior](https://en.wikipedia.org/wiki/Prior_probability#Informative_priors), [weakly informative prior](https://en.wikipedia.org/wiki/Prior_probability#Weakly_informative_priors), [uninformative priors](https://en.wikipedia.org/wiki/Prior_probability#Uninformative_priors), [imporper priors](https://en.wikipedia.org/wiki/Prior_probability#Improper_priors).)



### Example 1: Correct but not informative (relatively)
- **Assume the true p is 0.1.**
- **We believe that the true p is likely to be low. We use the Beta(2,6) to represent this information.**
- **It turns out that this information is correct, but the range is large.** 
- **The Bayes estimate and the Maximum likelihood estimate of p are both close to the true value.** 

```python
bayes_estimate, ml_estimate = Bayesian_ber_prior_beta(true_p=0.1, sample_size=50,
                                                      alpha_prior=2,beta_prior=6,plot=True)
print('Bayes Estimate:' + str(bayes_estimate))
print('Maximum Likelihood Estimate:' +str(ml_estimate))
```

**The codes below conduct 1000 experiments. In each experiments, the Bayes estimate and MLE are calculated and compared in term of their absolute bias. The mean absolute bias of the two methods are close.** 

```python
n_experiment = 1000
compare_bayes_mle = pd.DataFrame(columns=['n','bayes','mle','abs_bias_bayes','abs_bias_mle'])
for e in range(1,n_experiment+1):
    compare_bayes_mle.loc[e,'n'] = e
    bayes_estimate, ml_estimate = Bayesian_ber_prior_beta(true_p=0.1, sample_size=50,
                                                          alpha_prior=2,beta_prior=6,plot=False)
    compare_bayes_mle.loc[e,'bayes'] = bayes_estimate
    compare_bayes_mle.loc[e,'mle']            = ml_estimate
    compare_bayes_mle.loc[e,'abs_bias_bayes'] = abs(bayes_estimate-0.1)
    compare_bayes_mle.loc[e,'abs_bias_mle']   = abs(ml_estimate-0.1)
    
compare_bayes_mle[['abs_bias_bayes','abs_bias_mle']].mean()
```

## (2) Bayes estimate of Possion(lambda) with Exponential prior distribution

**The function below conducts an experimental analysis for Bayes estimate for Possion distribution with exponential prior distribution.** 

```python
def Bayesian_possion_prior_exp(true_Lambda,sample_size,Lambda_prior):
    sample = np.random.poisson(true_Lambda,sample_size)
    
    alpha_posterior = sample.sum()+ 1
    beta_posterior  = len(sample) + Lambda_prior
    
    x = np.linspace(0,10,1000)
    
    plt.plot(x,expon.pdf(x,scale=1/Lambda_prior),color='blue',
             label='prior distribution: Exp('+str(Lambda_prior)+')')
    plt.plot(x,gamma.pdf(x,a=alpha_posterior,scale=1/beta_posterior),color='red',
             label='posterior distribution: Gamma('+str(alpha_posterior)+','+str(beta_posterior)+')')
    plt.axvline(x=alpha_posterior/beta_posterior, color='red', linestyle='--', label='Bayes estimate')
    plt.axvline(x=sample.mean(), color='green', linestyle='--', label='MLE')
    plt.axvline(x=true_Lambda, color='black', linestyle='--', label='true value')
    plt.legend(bbox_to_anchor=(1,1))
    plt.show()

    return(round(alpha_posterior/beta_posterior,3),round(sample.mean(),3))
```

### Prior: Exp(0.5)

```python
bayes_estimate, ml_estimate = Bayesian_possion_prior_exp(true_Lambda=2, sample_size=50,
                                                         Lambda_prior=0.5)
print('Bayes Estimate:' + str(bayes_estimate))
print('Maximum Likelihood Estimate:' +str(ml_estimate))
```
