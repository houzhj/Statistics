**The codes below create random samples of X and Y following the given joint density function.**

```python
def q_x(p):
    return((-1+np.sqrt(1+8*p))/2)

def q_y_x(p,x):
    return(-x+np.sqrt(x**2+2*p*(x+0.5)))
```

```python
sample_size = 500
px   = np.random.uniform(0,1,sample_size)
py_x = np.random.uniform(0,1,sample_size)

x = [q_x(px[i]) for i in range(sample_size)]
y = [q_y_x(py_x[i],x[i]) for i in range(sample_size)]
```

**The codes below visualize the univariate distributions.**

```python
plt.hist(x,bins=20,density=True)
plt.title('Distribution of X')
plt.show()
plt.title('Distribution of Y')
plt.hist(y,bins=20,density=True)
plt.show()
```

**The codes below visualize the joint distributions.**

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y)
plt.show()
```

```python
df = pd.DataFrame(columns=['x','y'])
df['x'] = x
df['y'] = y
```

```python
sns.jointplot(data=df,x='x',y='y')
plt.show()
```

```python
sns.jointplot(data=df,x='x',y='y',kind='kde')
plt.show()
```
